```python
def removeConsecutiveDuplicates(s):
    if len(s)<2:
        return s
    if s[0]!=s[1]:
        return s[0]+removeConsecutiveDuplicates(s[1:])
    return removeConsecutiveDuplicates(s[1:])
```

```python
def normalize(text):
    text=text.lower().strip()
    text=text.replace("ì","i")
    text=text.replace("é","e")
    text=text.replace("š","s")
    text=text.replace("ḫ","h")
    text=text.replace("d","t")
    text=text.replace("/","")
    text=removeConsecutiveDuplicates(text)
    text=text.replace("ia","ya")
    return text
```

```python
def findinTags(query1):
    query=normalize("°"+str(query1)+"§")
    results=df[df.apply(lambda x: query in normalize(x['Tags_clean']), axis=1)]
    if len(results)>0:
        return results.index[0]
    else:
        return False
        
```

```python
def findinTagsD(query):
    result=findinTags(query)
    if "ia" in str(query):
        if result:
            return result
        else:
            query=query.replace("ia","iya")
            result=findinTags(query)
            return result
    else:
        return result
```
