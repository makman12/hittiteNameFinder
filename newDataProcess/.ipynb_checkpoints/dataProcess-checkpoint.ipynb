```python
import pandas as pd
import json
from math import nan, isnan
```

```python
names=pd.read_excel("Laman_all_names.xlsx")
deity=pd.read_excel("Laman_deity_name_instances.xlsx")
person=pd.read_excel("Laman_personal_name_instances.xlsx")
```

```python
names["Attestation"]=""
```

```python
names
```

```python
for i,row in names.iterrows():
    ins=row["Instances"]
    if type(ins)==str:
        ins=ins.split(", ")
        refs=[]
        for instance in ins:
            tip=instance[:3]
            kod=int(instance[3:])
            if tip=="DNI":
                ref=str(deity.iloc[kod]["Occurrence & notes"])
            elif tip=="PNI":
                ref=str(person.iloc[1000]["Attestation"])
            refs.append(ref)
        while "nan" in refs:
            refs.remove("nan")
        refs=", ".join(refs)
        names.at[i,"Attestation"]=refs
```
